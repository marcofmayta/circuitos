# 02 - Oraculos: Deutsch-Jozsa y Grover

Objetivo: estudiar como un oraculo codifica informacion en un circuito y como la interferencia permite extraerla.

```python
from quantum_circuits import deutsch_jozsa, grover_2qubits

constant = deutsch_jozsa("constant")
balanced = deutsch_jozsa("balanced")

print("Constante:", constant.operations, constant.probabilities())
print("Balanceada:", balanced.operations, balanced.probabilities())
```

En Deutsch-Jozsa de 1 bit, el primer qubit indica el tipo de funcion:

- primer qubit `0`: funcion constante.
- primer qubit `1`: funcion balanceada.

El segundo qubit es auxiliar y puede aparecer como `0` o `1`.

## Grover con 2 qubits

Grover empieza con superposicion uniforme, marca un estado con una fase negativa y luego usa el difusor para amplificarlo.

```python
for marked in ["00", "01", "10", "11"]:
    circuit = grover_2qubits(marked)
    print(marked, circuit.probabilities())
```

## Mirar las operaciones

El difusor usado aqui es `H^n X^n CZ X^n H^n`. Para 2 qubits, una iteracion basta para llevar toda la probabilidad al estado marcado.

```python
grover = grover_2qubits("10")
" -> ".join(grover.operations)
```
