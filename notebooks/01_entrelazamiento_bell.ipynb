# 01 - Entrelazamiento y estados de Bell

Objetivo: construir un estado de Bell y observar que sus resultados estan correlacionados.

```python
from quantum_circuits import QuantumCircuit, bell_state

bell = bell_state()
print(bell.operations)
print(bell.amplitudes())
print(bell.probabilities())
```

El circuito usa dos pasos:

1. `H(0)` crea superposicion en el primer qubit.
2. `CNOT(0 -> 1)` copia la correlacion al segundo qubit sin clonar un estado arbitrario.

El resultado teorico es `( |00> + |11> ) / sqrt(2)`.

```python
manual = QuantumCircuit(2)
manual.h(0)
manual.cnot(0, 1)
manual.probabilities()
```

## Muestreo

Aunque cada medicion individual es incierta, los resultados validos son solo `00` y `11`.

```python
bell.sample(shots=1000, seed=42)
```

## Variantes

Puedes generar otros estados de Bell cambiando fases o aplicando `X` antes/despues del `CNOT`.

```python
variant = QuantumCircuit(2).x(0).h(0).cnot(0, 1)
variant.amplitudes(), variant.probabilities()
```
