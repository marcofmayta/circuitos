# 00 - Fundamentos de circuitos cuanticos

Objetivo: entender como representamos qubits, amplitudes, probabilidades y compuertas usando el simulador del proyecto.

En este material usamos una convencion **big-endian**: el qubit 0 es el bit mas significativo. Por ejemplo, en `|10>` el qubit 0 vale 1 y el qubit 1 vale 0.

```python
from quantum_circuits import QuantumCircuit

qc = QuantumCircuit(1)
qc.amplitudes(), qc.probabilities()
```

## Superposicion con Hadamard

La compuerta `H` transforma `|0>` en una combinacion equilibrada de `|0>` y `|1>`. Las probabilidades son iguales, aunque las amplitudes contienen la informacion completa del estado.

```python
qc = QuantumCircuit(1).h(0)
print(qc.operations)
print(qc.amplitudes())
print(qc.probabilities())
```

## Medicion simulada

Una medicion no devuelve amplitudes: devuelve resultados clasicos. Al repetir muchas veces, las frecuencias se acercan a las probabilidades teoricas.

```python
qc.sample(shots=20, seed=1), qc.sample(shots=1000, seed=1)
```

## Dos qubits

Con dos qubits tenemos cuatro estados base: `|00>`, `|01>`, `|10>` y `|11>`. Aplicar Hadamard a ambos crea una superposicion uniforme.

```python
qc2 = QuantumCircuit(2).h(0).h(1)
qc2.probabilities()
```
