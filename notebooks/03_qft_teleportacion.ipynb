# 03 - QFT y teleportacion cuantica

Objetivo: explorar dos circuitos centrales: la Transformada Cuantica de Fourier y la base del protocolo de teleportacion.

## Transformada Cuantica de Fourier

La QFT transforma informacion de la base computacional en fases. Es una pieza clave de algoritmos como estimacion de fase y Shor.

```python
from quantum_circuits import qft, teleportation

circuit = qft(qubits=3, basis_state=1)
print("Operaciones:", " -> ".join(circuit.operations))
circuit.amplitudes(decimals=4)
```

La magnitud de las amplitudes queda equilibrada, pero las fases cambian. Por eso mirar solo probabilidades no cuenta toda la historia.

```python
qft(qubits=3, basis_state=1).probabilities()
```

## Teleportacion cuantica

La teleportacion no mueve materia: transfiere el estado de un qubit usando entrelazamiento, medicion y comunicacion clasica. Aqui construimos la parte unitaria previa a la correccion clasica.

```python
tele = teleportation(alpha=1, beta=0)
print("Operaciones:", " -> ".join(tele.operations))
tele.probabilities()
```

Prueba un estado distinto, por ejemplo una superposicion normalizada `( |0> + |1> ) / sqrt(2)`.

```python
import numpy as np

tele_super = teleportation(alpha=1 / np.sqrt(2), beta=1 / np.sqrt(2))
tele_super.amplitudes(decimals=4)
```
